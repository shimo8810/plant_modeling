```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('./p2b_log', 'r') as f:
    jf = json.load(f)
dec_losses = []
enc_losses = []
dis_losses = []
iters = []
for snap in jf:
    iters.append(snap['iteration'])
    dec_losses.append(snap['dec/loss'])
    enc_losses.append(snap['enc/loss'])
    dis_losses.append(snap['dis/loss'])
```

```python
plt.plot(iters, dec_losses)
plt.plot(iters, np.array(dis_losses))
plt.grid()
plt.show()
```
